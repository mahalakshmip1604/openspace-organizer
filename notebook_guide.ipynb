## Step 1: Build a Seat

Create a class called `Seat` with two attributes:

- `free` which is a boolean.
- `occupant` which is a string.

and 2 methods (functions): 

- `set_occupant(name)` which allows the program to assign someone a seat if it's free
- `remove_occupant()` which  remove someone from a seat and return the name of the person occupying the seat before


```python
class Seat:
    def __init__(self, free, occupant=None):
        self.free = free 
        self.occupant = occupant

    def set_occupant(self, name):
        # Check if the seat is free before assigning an occupant
        if self.free:
            self.occupant = name
            self.free = False
            print(f"{name} has been assigned to the seat.")
        else:
            print(f"Sorry, the seat is already occupied by {self.occupant}.")

    def remove_occupant(self):
            # Check if the seat is occupied before removing the occupant
            if not self.free:
                removed_person = self.occupant
                self.occupant = None
                self.free = True
                return removed_person
            else:
                 print("The seat is already free.")
                 return None 
```

```python

my_seat = Seat(free=True, occupant="Mahalakshmi")

my_seat.set_occupant("Mahalakshmi")
print(my_seat.occupant)  # Output: Mahalakshmi

my_seat.remove_occupant()
print(my_seat.occupant)  # Output: None
```

## Step 2: Build a Table

Create a class `Table` with ? attributes:

- `capacity` which is an integer
- `seats` which is a list of `Seat` objects (size = `capacity`)

and 3 methods (functions): 
- `has_free_spot()` that returns a boolean (True if a spot is available)
- `assign_seat(name)` that places someone at the table
- `left_capacity()` that returns an integer

Question: Which attributes make sense to give? For now let's say we want to build 6 tables with 4 seats.


```python

class Table:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.seats = [Seat(True, occupant="") for i in range(capacity)]

    def has_free_spot(self):
        for seat in self.seats:
            if seat.free:
                return True
        return False

    def assign_seat(self, name):
        # Check if there is a free spot before trying to assign a seat
        for seat in self.seats:
            if seat.free:
                seat.set_occupant(name)
                return True
        print("Sorry, no free seats available.")

    def left_capacity(self):
        #counts and returns the number of free seats left at the table
        count = 0
        for seat in self.seats:           
            if seat.free:
                count += 1
        return count   
```

```python

test_table = Table(capacity=6)
print(test_table.has_free_spot())  # Output: True
test_table.assign_seat("Mahalakshmi")
print(test_table.left_capacity())  # Output: 5



    
```

And some methods:
